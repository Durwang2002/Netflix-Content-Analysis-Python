```python
netflix.isnull().sum()
```

```python
# Handling Missing Values
```

```python
netflix["director"].fillna("Unknown", inplace=True)
```

```python
netflix["cast"].fillna("Unknown", inplace=True)
```

```python
netflix["country"].fillna("Unknown", inplace=True)
```

```python
netflix.isnull().sum()
```

```python
#Fill rating using mode:
```

```python
rating_mode = netflix["rating"].mode()[0]
netflix["rating"].fillna(rating_mode,inplace=True)
```

```python
#Remove missing date records
```

```python
netflix.dropna(subset=["date_added"],inplace=True)
```

```python
#Verify Missing Values
```

```python
netflix.isnull().sum()
```

```python
#Remove Duplicate Records
```

```python
netflix.duplicated().sum()
```

```python
#Verify Data Types
```

```python
netflix.info()
```

```python
#STEP 3: Feature Engineering
```

```python
#Extract Date Features
#Create Year, Month, Day
```

```python
netflix["added_year"] = (netflix["date_added"].dt.year)
```

```python
netflix["added_month"] = (netflix["date_added"].dt.month_name())
```

```python
netflix["added_day"] = (netflix["date_added"].dt.day_name())
```

```python
#Calculate Content Age
```

```python
current_year = datetime.now().year

netflix["content_age"] = (
    current_year - netflix["release_year"]
)
```

```python
#Check Columns
```

```python
netflix.head()
```

```python
#Separate Movie Duration and TV Seasons
#Extract Numeric Value
```

```python
netflix["duration_value"] = (netflix["duration"].str.extract(r"(\d+)").astype(float))
```

```python
#Create Duration Unit
```

```python
netflix["duration_unit"] = (netflix["duration"].str.extract(r"([A-Za-z]+)"))
```

```python
#Check Result
```

```python
netflix.head()
```

```python
#Create Movie Duration Categories
```

```python
movie_condition = (netflix["type"] == "Movie")
```

```python
netflix.loc[movie_condition & (netflix["duration_value"] < 60),"duration_category"] = "Short Movie"
```

```python
netflix.loc[ movie_condition & (netflix["duration_value"] >= 60) & (netflix["duration_value"] <= 120), "duration_category"] = "Medium Movie"
```

```python
netflix.loc[movie_condition & (netflix["duration_value"] > 120),"duration_category"] = "Long Movie"
```

```python
#IQR Method
```

```python
Q1 = movies["duration_value"].quantile(0.25)

Q3 = movies["duration_value"].quantile(0.75)

IQR = Q3 - Q1


lower_limit = Q1 - 1.5 * IQR

upper_limit = Q3 + 1.5 * IQR


outliers = movies[
    (movies["duration_value"] < lower_limit) |
    (movies["duration_value"] > upper_limit)
]


print(
    "Total Outliers:",
    len(outliers)
)
```

```python
#Distribution Analysis
```

```python
#Skewness Analysis

#Skewness = 0     → Symmetric

#Skewness > 0     → Right Skewed

#Skewness < 0     → Left Skewed
```

```python
movie_duration = movies["duration_value"]

skewness = movie_duration.skew()

print("Skewness:", round(skewness,3))
```

```python
#Kurtosis Analysis

#Kurtosis > 3 Heavy tails

#Kurtosis < 3 Light tails

#Kurtosis = 3 Normal distribution
```

```python
kurtosis = movie_duration.kurt()

print("Kurtosis:", round(kurtosis,3))
```

```python
#Normality Testing
```

```python
#Shapiro-Wilk Test
```

```python
from scipy.stats import shapiro

sample_data = movie_duration.sample(min(5000, len(movie_duration)),random_state=42)

stat, p_value = shapiro(sample_data)

print("Statistic:", stat)
print("P-value:", p_value)
```

```python
#Decision
```

```python
alpha = 0.05

if p_value > alpha:

    print("Data appears normally distributed")

else:

    print("Data is NOT normally distributed")
```

```python
#Confidence Interval Analysis
```

```python
from scipy.stats import t

mean_duration = movie_duration.mean()

std_duration = movie_duration.std()

n = len(movie_duration)

confidence = 0.95

margin_error = t.ppf((1 + confidence)/2,n-1) * (std_duration / np.sqrt(n))

lower_bound = (mean_duration - margin_error)

upper_bound = (mean_duration + margin_error)

print(
    f"95% Confidence Interval: "
    f"({lower_bound:.2f}, {upper_bound:.2f})"
)
```
